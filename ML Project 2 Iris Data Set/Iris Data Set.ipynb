```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris = pd.read_csv("iris.data") 
print(iris.shape)
print(iris.columns)
iris
```

```python
iris['species'].value_counts()
```

```python
Iris_setosa = iris.loc[iris['species'] == 'Iris-setosa']
Iris_versicolor = iris.loc[iris['species'] == 'Iris-versicolor']
Iris_virginica = iris.loc[iris['species'] == 'Iris-virginica']
```

```python
Iris_setosa.plot.scatter(x='sepal_length', y='petal_length', color='DarkGreen', label='Iris_setosa')
Iris_versicolor.plot.scatter(x='sepal_length', y='petal_length', color='Blue', label='Iris_versicolor')
Iris_virginica.plot.scatter(x='sepal_length', y='petal_length', color='Red', label='Iris_virginica')
```

```python
iris.plot.scatter(x='sepal_length', y='sepal_width')
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue="species", size = 4).map(plt.scatter, "sepal_length", "sepal_width").add_legend();
```

```python
sns.FacetGrid(iris, hue="species", size = 4).map(plt.scatter, "petal_length", "petal_width").add_legend();
```

```python
sns.pairplot(iris, hue= 'species', size=3 , diag_kind= 'kde')
```

```python
x = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = iris['species']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.2)
knn = KNeighborsClassifier(n_neighbors= 1)
knn.fit(X_train, y_train)
```

```python
X_new =np.array([[4.0, 1.5, 1.0, .5]])
```

```python
#test set accuracy
print(knn.score(X_test, y_test))
```

```python
#prediction of new one
print(knn.predict(X_new))
```
