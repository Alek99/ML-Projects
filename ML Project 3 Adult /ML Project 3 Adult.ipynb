```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
adult = pd.read_csv("adult.data") 
print(adult.shape)
print(adult.columns)
type(adult)
```

```python
adult['income']=adult['income'].map({' <=50K':1,' >50K':0}).astype(int)
```

```python
adult['sex']=adult['sex'].map({' Male':1,' Female':0}).astype(int)
```

```python
mapp={' White':0,' Amer-Indian-Eskimo':1,' Asian-Pac-Islander':2,' Black':3,' Other':4}
adult['race']= adult['race'].map(mapp)
    
```

```python
mapp = {' Unmarried':0,' Wife':1,' Husband':2,' Not-in-family':3,' Own-child':4,' Other-relative':5, ' ?':6}
adult['relationship'] =  adult['relationship'].map(mapp)
```

```python
mapp = {' Married-civ-spouse':0,' Divorced':1,' Never-married':2,' Separated':3,' Widowed':4,' Married-spouse-absent':5, ' Married-AF-spouse':6, ' ?':7}
adult['marital-status'] =  adult['marital-status'].map(mapp)
```

```python
employment_map = {' Federal-gov':1,' Local-gov':1,' Private':3, ' Self-emp-inc':2,' Self-emp-not-inc':2,' State-gov':1,' Without-pay':0, ' ?':0, ' Never-worked':0}
adult['workclass'] = adult['workclass'].map(employment_map)
```

```python
mapp = {' Tech-support':0, ' Craft-repair':1, ' Other-service':2, ' Sales':3, ' Exec-managerial':4, ' Prof-specialty':5, ' Handlers-cleaners':6, ' Machine-op-inspct':7, ' Adm-clerical':8, ' Farming-fishing':9, ' Transport-moving':10, ' Priv-house-serv':11, ' Protective-serv':12, ' Armed-Forces':13, ' ?':14}
adult['occupation'] =  adult['occupation'].map(mapp)
```

```python
country = np.array([])
for dataset in adult['native-country']:
    if dataset != ' United-States':
        dataset = 1
    if dataset == ' United-States':
        dataset = 0

    country = np.append(country, [dataset])
print(country)

adult['native-country'] = country
```

```python
adult.drop(labels=['education-num'],axis=1,inplace=True)
```

```python
adult.loc[(adult['capital-gain'] > 0),'capital-gain'] = 1
adult.loc[(adult['capital-gain'] == 0 ,'capital-gain')]= 0
adult.loc[(adult['capital-loss'] > 0),'capital-gain'] = 1
adult.loc[(adult['capital-loss'] == 0 ,'capital-gain')]= 0
```

```python
print(adult.columns)
adult.isnull().values.any()
```

```python
adult.isnull().sum()
```

```python
x = adult[['age', 'workclass', 'fnlwgt',
       'marital-status', 'occupation', 'relationship', 'race', 'sex',
       'capital-gain', 'capital-loss', 'hours-per-week', 'native-country']]
y = adult['income']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.3)
X_train, X_val, y_train, y_val = train_test_split(X_train,y_train,test_size=0.2,random_state=5)

X_train
```

```python
print ("Train dataset: {0}{1}".format(X_train.shape, y_train.shape))
print ("Test dataset: {0}{1}".format(X_test.shape, y_test.shape))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn import model_selection
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
models = []
names = ['LR','Random Forest','Neural Network','GaussianNB','DecisionTreeClassifier','SVM',]

models.append((LogisticRegression()))
models.append((RandomForestClassifier(n_estimators=100)))
models.append((MLPClassifier()))
models.append((GaussianNB()))
models.append((DecisionTreeClassifier()))
```

```python
kfold = model_selection.KFold(n_splits=5,random_state=7)

for i in range(0,len(models)):    
    cv_result = model_selection.cross_val_score(models[i],X_train,y_train,cv=kfold,scoring='accuracy')
    score=models[i].fit(X_train,y_train)
    prediction = models[i].predict(X_val)
    acc_score = accuracy_score(y_val,prediction)     
    print ('-'*40)
    print ('{0}: {1}'.format(names[i],acc_score))
```

```python
randomForest = RandomForestClassifier(n_estimators=100)
randomForest.fit(X_train,y_train)
prediction = randomForest.predict(X_test)
```

```python
print ('-'*40)
print ('Accuracy score:')
print (accuracy_score(y_test,prediction))
print ('-'*40)
print ('Confusion Matrix:')
print (confusion_matrix(y_test,prediction))
print ('-'*40)
print ('Classification Matrix:')
print (classification_report(y_test,prediction))
```
